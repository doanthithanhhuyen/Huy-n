```python
import datetime

def dict_from_entries(keys, values):
    new_dict = {}
    for i in range(len(keys)):
        new_dict[keys[i]] = values[i]
    return new_dict

def safe_input(input_type, message):
    while True:
        value = input(message).strip()
        if value:
            if input_type == 'string':
                return value
            elif input_type == 'int_positive' and value.isnumeric():
                return int(value)
            elif input_type == 'float_positive' and value.replace('.', '').isnumeric():
                return float(value)
            
            
def find_key(dict_list, key, value):
    for e in dict_list:
        if str(e[key]).upper() == str(value).upper():
            return e


def select_id_or_name(dict_list, name):
    while True:
        selected = None
        name_or_id = safe_input('string', 'Name or Id: ')
        print()

        if name_or_id.isnumeric():  #isdecimal: trả về True nếu chuỗi ký tự chỉ chứa các ký tự thập phân
            selected = find_key(dict_list, 'id', int(name_or_id))
        else:
            selected = find_key(dict_list, 'name', name_or_id)

        if selected:
            return selected
        else:
            print("No %s with that criteria." % name)
            print("Try again")


def today():
    date_today = datetime.datetime.now()
    return date_today.strftime("%x")


def dict_to_csv_line(dictionary, keys):
    line = ''
    for k in keys:
        line += str(dictionary[k])
        if k != keys[-1]:
            line += ','
        else:
            line += '\n'
    return line

def get_products():
    data = []
    file = open('products.csv')

  #  search = input('ID or name of the product?: ')
    for line in file:
        spacing = line.split(',')
        product = {
            "id": int(spacing[0]),
            "name": spacing[1],
            "price": float(spacing[2]),
            "quantity": int(spacing[3]),
            "type": spacing[4],
            "sub_type": spacing[5]
        }
        data.append(product)

    file.close()
    return data


def update_products(product_list):
    file = open('products.csv')
    for p in product_list:
        def dict_to_csv_line(dictionary, keys):
            line = ''
            for k in keys:
                line += str(dictionary[k])
                if k != keys[-1]:
                    line += ','
                else:
                    line += '\n'
            return line 
    
    file.close()


columns = (
    'id',
    'total_price',
    'num_products',
    'product_id',
)


def get_sales():
    sales = []
    file = open('sales.csv')
    for line in file:
        spacing = line.split(',')
        sale = {
            "id": int(spacing[0]),
            "total_price": float(spacing[1]),
            "num_products": int(spacing[2]),
            "product_id": int(spacing[3]),
        }
        sales.append(sale)

    file.close()
    return sales


def add_sale(sale):
    file = open('sales.csv', 'a')
    line = dict_to_csv_line(sale, columns)

    file.write(line)
    file.close()



def register_product_arrival():
    database = ('Huyền', "123456")
    username = input("Enter Your Username : ")
    password = input("Enter Your Password : ")
    if username != database[0] and password != database[1]:
        print("ERROR! TRY IT!\n")
    else: 
        print("--- Cập nhập sản phẩm ---")
        print("Đó là sản phẩm nào")
        menu = get_products()

        for thing in menu:
            print(thing['id'], thing['name'])

        chooosen_product = select_id_or_name(menu, 'product')

        arrival_quantity = safe_input('int_positive', 'Số lượng hàng nhập thêm:')

        chooosen_product['quantity'] += arrival_quantity

        print('Mạt hàng %s vừa được thêm vào kho là %s sản phẩm' %
              (chooosen_product['name'], arrival_quantity))
        print('Hiện tại còn %s' % chooosen_product['quantity'])
        update_products(menu)
    
    
def query_data():

    print("---Thông tin sản phẩm---")
    print("Chọn sản phẩm")

    list_products = get_products()

    for products in list_products:
        print(products['name'], products['id'])

    this_product = select_id_or_name(list_products, 'product')

    print("id:", this_product["id"])
    print("Price per unit:", this_product["price"])
    print("Quantity in stock:", this_product["quantity"])
    print("Categories:", this_product["type"], "-", this_product["sub_type"])
    
def register_sale():
    sale = {
        "date": today()
    }
    print("--- Đặt hàng ---\n")

    products = get_products()

    for p in products:
        print("- %s (%s) (%s in stock)" %
              (p['id'], p['name'], p['quantity']))
    print()

    product = select_id_or_name(products, 'product')

    sale['product_id'] = product['id']

    print("Selected: %s (%s) (%s in stock)" %
          (product['name'], product['id'], (product['quantity'])))

    quantity = 0
    while True:
        quantity = safe_input("int_positive", "Nhập số lượng bạn muốn đặt. ")
        if quantity > 0 and quantity <= product['quantity']:
            print("Vui lòng đợi nhân viên chuẩn bị hàng cho bạn!")
            break
        else:
            print(
                "Không hợp lệ. Vui lòng chọn số lượng không lớn hơn số lượng trong kho.")

    
    product['quantity'] = quantity
    sale['num_products'] = quantity
    sale['total_price'] = quantity * product['price']

    print("\nTổng thanh toán: $%s " %
          (sale['total_price']))

    sale['id'] = len(get_sales())

    print("\nId đơn hàng của bạn là: ", sale['id'])

    update_products(products)
    add_sale(sale)


def main():     
    print("HỆ THỐNG BÁN HÀNG VÀ HÀNG TỒN KHO")
    print("""
Select an action
1. Đặt hàng
2. Truy cập sản phẩm
3. Cập nhập số lượng của các mặt hàng

""")

    while True:
         
        action = safe_input('int_positive', 'Action: ')

        if action > 0 and action < 4:
            break

        print("Oops! Thử lại với các action từ 1 đến 3")
    if action == 1:
        register_sale()
    elif action == 2:
        query_data()
    elif action == 3:
        register_product_arrival()
    
    print("Cảm ơn bạn đã sử dụng hệ thống. Chúc bạn một ngày tốt lành!")

while True:
    main()

    again = input("Bạn có muốn bắt đầu lại? (y/n) ").strip()[0].lower()

    if (again == 'y'):
        continue
    else:
        print("See you later!\n")
        break
```
