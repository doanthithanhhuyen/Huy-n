```python
import turtle
tl = turtle.Turtle()
tl.shape("turtle")
screen = tl.getscreen()
screen.bgcolor('black')
tl.hideturtle()
tl.pencolor('gray')
tl.speed(100)
#
tl.penup()
tl.goto(0,-120)
tl.pendown()
tl.fillcolor('white')
tl.begin_fill()
tl.circle(120,360)
tl.end_fill()
#
tl.penup()
tl.goto(0,100)
tl.pendown()
tl.fillcolor('white')
tl.begin_fill()
tl.circle(60,360)
tl.end_fill()
tl.penup()
tl.goto(-30,160)
tl.pendown()
tl.fillcolor('red')
tl.begin_fill()
tl.circle(5,360)
tl.penup()
tl.goto(30,160)
tl.pendown()
tl.circle(5,360)
tl.end_fill()

tl.penup()
tl.goto(0,150)
tl.pendown()
tl.fillcolor('yellow')
tl.begin_fill()
tl.circle(7)
tl.end_fill()
#vẽ mũ
tl.penup()
tl.goto(-60,200)
tl.pensize(20)
tl.pendown()
tl.forward(120)

tl.penup()
tl.goto(-42,215)
tl.pendown()
tl.pensize(23)
tl.left(90)
tl.forward(20)
tl.penup()
tl.goto(42,215)
tl.pendown()
tl.forward(20)

tl.penup()
tl.goto(-50,215)
tl.pendown()
tl.right(90)
tl.pensize(15)
tl.pencolor('red')
tl.forward(100)
tl.penup()
tl.goto(-25,250)
tl.pendown()
tl.pencolor('gray')
tl.pensize(55)
tl.forward(50)

#vẽ khuy áo
tl.pencolor('gray')
tl.pensize(5)
tl.penup()
tl.goto(0,50)
tl.pendown()
tl.fillcolor('gray')
tl.begin_fill()
tl.circle(7,360)
tl.penup()
tl.goto(0,0)
tl.pendown()
tl.circle(7,360)
tl.end_fill()

tl.penup()
tl.goto(-20,130)
tl.pendown()
tl.right(30)
tl.forward(20)
tl.left(60)
tl.forward(20)
tl.right(30)
#vẽ khăn

tl.penup()
tl.goto(-50,107)
tl.pendown()
tl.pensize(18)
tl.pencolor('red')
tl.forward(100)
tl.right(90)
tl.forward(30)
tl.penup()
tl.goto(32,107)
tl.pendown()
tl.forward(20)
tl.left(90)


tl.penup()
tl.goto(90,10)
tl.pencolor('gray')
tl.pensize(5)
tl.pendown()
tl.left(30)
tl.forward(60)
tl.penup()
tl.left(180)
tl.forward(10)
tl.left(130)
tl.pendown()
tl.forward(10)
tl.penup()
tl.left(180)
tl.forward(10)
tl.pendown()
tl.right(90)
tl.forward(15)
tl.right(60)
tl.penup()
tl.goto(-90,10)
tl.pendown()
tl.left(130)
tl.forward(60)
tl.penup()
tl.left(180)
tl.forward(10)
tl.pendown()
tl.left(130)
tl.forward(10)
tl.penup()
tl.left(180)
tl.forward(10)
tl.right(90)
tl.pendown()
tl.forward(15)

turtle.done()
```
